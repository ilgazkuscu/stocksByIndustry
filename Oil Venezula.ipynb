```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Get the stock data
oil_comppanies = ['CVX', 'XOM', 'COP']

df = yf.download(oil_comppanies, start='2025-12-17', end='2026-02-04', auto_adjust=False)

df = pd.DataFrame(df.round(2))

df['Adj Close'].tail()
```
